# NumPy Kütüphanesi

- NumPy(Numerical Python) çok boyutlu dizilerle(array) ve matrislerle işlem yapmamıza olanak sağlayan yüksek hacimli matematiksel işlemleri yapmak için temel bir kütüphanedir.
- NumPy'ın temelini NumPy dizileri(array) oluşturur ve temel python'daki listelere benzer.
- Python listelerinden farklı olarak NumPy dizileri homojen yapıda olmalıdır. dizi içindeki tüm elemanlar **_aynı veri tipinden_** olmalıdır.
- =>numpy array=( [0,1,2,3,4,5,6,7] 
- Numpy paketinin temelinde **_ndarray_** (N-dimensional array) nesnesi vardır. n=>çok boyutlu olmayı gösterir
- Bir Numpy dizisi 1-boyutlu, 2-boyutlu, 3-boyutlu... yani çok boyutlu veriden oluşabilir.
- Numpy dizisine koyduğumuz veri **_"tek"_** tipten oluşmalıdır.(Ya tamsayıi ya ondalıklı, ya karakter..vs.)


## **_NumPy Kütüphanesi İmport Etme_**

```python
import numpy as np
```

## **_NumPy ile Temel İşlemler_**

```python
#Temel Python listesi oluşturalım
data1=[6,7.5,8,0,1]
```

```python
#Array'a dönüştürelim
dizi1=np.array(data1)
dizi1
```

### **_NOT:_** 
NumPy dizisine çevirdiğimiz data1 listesi->dizi1 değişkenine atandıktan sonra çıktı[OUT] olarak **_float_** veri tipi ile dönmüştür. Burada NumPy kütüphanesinin tek tip veri tipi üzerinden işlem yaptığı görülebilir. **_Veri kaybı olmaması için tüm sayıları temsil edecek veri tipini bulur ve dönüştürür._**
- Bu örnekte eğer tamsayı olarak dönseydi 7.5 içerisindeki 0.5'i yuvarlamak durumunda kalırdı ki bu da veri kaybına sebep olurdu.

```python
#dizi1 değikeninin tipini bulalım.
type(dizi1)
```

```python
# dizi1 dizisinin içendeki verinin tipini bulalım
dizi1.dtype
```

### NOT
Temel Python'da "nesnelerin" iki temel özelliği vardır:
- method=fonksiyonlar
- attribute=nesnelerde tuttuğumuz blgiler, verile

NumPy'da bilgilerin array'da nasıl tutulduğunu gösteren bir attribute tanımlanmıştır. Yani elimizdeki ndarray'in içindeki veri tipini görebiliriz. Bunun attribute'u **_.dtype_** komutudur.

### "type" ile "dtype" arasındaki fark

- type; bize bir nesnenin genel tipini gösterir.(hangi sınıfa ait ...gibi)
- dtype ise; vaeolan bir nesnenin içinde o verilen nasıl tutulmuştur(tamsayı mı, karakter mi, float mı...gibi)
